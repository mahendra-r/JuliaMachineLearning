```julia
using PyPlot

n = 50
srand(1)
x = rand(n)
y = rand(n)
area = pi .* (15 .* rand(n)).^2 # 0 to 15 point radiuses
scatter(x, y, s=area, alpha=0.5)
```

```julia
alphas = [0.0, 0.8, 0.98]
T = 200
x = zeros(T + 1)

for alpha in alphas
    x[1] = 0
    for t in 1:T
        x[t+1] = alpha * x[t] + randn()
    end
    plot(x, label="alpha = $alpha")
end
legend()
```

```julia
function generate_data(n)
    epsilon_values = Array(Float64, n)
    for i = 1:n
        epsilon_values[i] = randn()
    end
    return epsilon_values
end

ts_length = 100
data = generate_data(ts_length)
plot(data, "b-")
```

```julia
brightness(color) = color.r + color.g + color.b
brightness(color::Colors.AlphaColor) = color.c.r + color.c.g + color.c.b

function getvalue(img, i, j)
    w, h = size(img)
    if i < 1 || j < 1 || i > w || j > h
        return 0.0f0
    end
    return brightness(img[i, j])
end

function energy(img, i, j)
    # A B C
    # D E F
    # G H I
    w, h = size(img)
    a, d, g, b, e, h, c, f, i = [getvalue(img, x, y) for x=i-1:i+1, y=j-1:j+1]
    xenergy=a + 2d + g - c - 2f - i
    yenergy=a + 2b + c - g - 2h - i
    √((xenergy^2) + (yenergy^2))
end
```
